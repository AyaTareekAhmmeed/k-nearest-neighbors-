```python
#importing libraries
import numpy as np
import pandas as pd
```

```python
#loading data
dataset = pd.read_csv('houses.csv')

#showing data
dataset.head(20)
```

```python
#cleaning data "changing nan"
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp = imp.fit(dataset)
dataset = imp.transform(dataset)
```

```python
#X Data input
X = dataset[:, :-1]
print('X Data is \n' , X[:10])
print('X shape is ' , X.shape)

#y Data output
y = dataset[:, -1]
print('y Data is \n' , y[:10])
print('y shape is ' , y.shape)
```

```python
#Scaling data "rearanging data"
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
print("X after rearanging data\n", X)
```

```python
#Splitting data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

#Splitted Data
print('X_train shape is ' , X_train.shape)
print('X_test shape is ' , X_test.shape)
print('y_train shape is ' , y_train.shape)
print('y_test shape is ' , y_test.shape)
```

```python
#Applying KNeighborsRegressor Model
'''
#sklearn.neighbors.KNeighborsRegressor(n_neighbors=5, weights=, algorithm=’auto’, leaf_size=30, #n_neighbors= num of used neighbor 
#   p=2, metric=’minkowski’, metric_params=None,n_jobs=None) #weights=distance, or defined function or uniform used to calulate weight
                                                             #algorithm=’auto or ball_tree ,  kd_tree  , brute

'''

from sklearn import neighbors #K Neighbors Regressor
knn = neighbors.KNeighborsRegressor(n_neighbors = 4)
y_ = knn.fit(X_train, y_train)

#Calculating Details
print('KNeighborsRegressorModel Train Score is : ' , y_.score(X_train, y_train)) #accurancy for training data
print('KNeighborsRegressorModel Test Score is : ' , y_.score(X_test, y_test)) #accurancy foor testing
```

```python
#Calculating Prediction
y_pred = knn.predict(X_test) 
print('Predicted Value for KNeighborsRegressorModel is : \n' , y_pred[:10])

#calcylating actual output
print('Actual Value for KNeighborsRegressorModel is : \n' , y_test[:10])
```

```python
#calculating errors for regression model
#Calculating Mean Absolute Error
from sklearn.metrics import mean_absolute_error 
MAEValue = mean_absolute_error(y_test, y_pred, multioutput='uniform_average') # it can be raw_values
print('Mean Absolute Error Value is : ', MAEValue)

#----------------------------------------------------
#Calculating Mean Squared Error
from sklearn.metrics import mean_squared_error
MSEValue = mean_squared_error(y_test, y_pred, multioutput='uniform_average') # it can be raw_values
print('Mean Squared Error Value is : ', MSEValue)

#----------------------------------------------------
#Calculating Median Squared Error
from sklearn.metrics import median_absolute_error
MdSEValue = median_absolute_error(y_test, y_pred)
print('Median Squared Error Value is : ', MdSEValue )
```
